```python
import torch
state_dict = torch.load("./model.pth")
```

```python
{name: param.shape for name, param in state_dict.items()}
```

```python
dataset = torch.load("./dataset.pt")
```

```python
list(map(lambda ls: torch.stack(ls),(zip(*dataset["eval"]))))[1]
```

```python
from train_transformer import eval_epoch
from torch.utils.data import DataLoader
```

```python
from miexp.models.interptransformer import SingleHeadTransformer


model = SingleHeadTransformer(vocab_size=2, head_dim=8, hidden_dim=8)
model.load_state_dict(state_dict)
```

```python
type(model)
```

```python
eval_dl = DataLoader(dataset["eval"], batch_size=32, shuffle=False)
res = eval_epoch(model, eval_dl, device="mps")
```
